# 1. Random Number Generation.

```python
X = np.random.rand(1000, 1)               # from uniform distribution over [0, 1), shape = (100, 1).
y = 2*X + 3 + np.random.randn(1000, 1)    # from normal distribution, i.e. N(0, 1), shape = (100, 1). create Gaussian noise.

plt.scatter(X, y)
```
